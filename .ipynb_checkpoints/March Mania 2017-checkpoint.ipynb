# March Mashine Learning Mania 2017
The objective is to predict the results of the NCAA tournament for the year 2017. For all possible matchups that can occur a predicted win probability is computed for the team with lower teamID to win. Then the total score for the competition is computed using the Log_Loss function:
$$ {\rm Log\_Loss} = -\frac{1}{n} \sum_{i=1}^n \left(y_i \log \hat{y}_i + (1-y_i)\log( 1-\hat{y}_i) \right) $$
Here $y_i$ is the actual outcome of game $i$ (i.e., equal to either $0$ or $1$) and $\hat{y}_i$ is the predicted outcome. A perfect score is zero. The Log_Loss is the same as $1/n$ times the log of the predicted probability for the outcome of all games that actually occured, treating the games as independent events.

## Exploratry Data Analysis

The goal of this exploratory analysis will be to identify those features which are most correlated with winning games. Ultimately we have to use historical data to make predictions about game outcomes, so we are mostly looking for correlations between the **average** stats of a team and their probability of winning.

We will be guided by the principle that **efficiencies** and **ratios** are the preferred variables to look at.

Now we define a few functions which will compile statistics for us.

-team_stats takes as input a team_ID and (optionally) a season and outputs all of the stastistics we plan to use for that team (for that season), game by game.

-matchup_stats takes as input two team_IDs and a season and outputs the mean stats of each of those two teams in that season.

-tourn_stats takes as input a season and outputs all of the matchup stats for each game played in that tournament, together with a record of which team won the game. Also included are the seed numbers for each team.

These three functions build on each other, and the output of tourn_stats for a range of seasons will serve as the training set for our model.

```python
#raw_features are the per-game stats we use to evaluate the teams.
raw_features = ['fge','fge3','fte','to','blk','stl','pf','ast','score', 'dr','or']

def team_stats(team_ID, Season = 0):
    ''' 
    Input: team_ID number and (optionally) a year for Season
    Output: Dataframs with the game-by-game statistics of a given team for the given Season or
        (when no season is specified) all seasons.

    The statistics which are computed are the efficiencies (i.e., baskets/attempts), and for each game
    we record whether the teams was the winner or loser.
    '''   
    
    X = results_detailed
    
    #Compute per-game efficiency stats
    #Field Goal efficiency
    X['Wfge'] = X['Wfgm']/X['Wfga']
    X['Lfge'] = X['Lfgm']/X['Lfga']
    #Three-pointer efficiency
    X['Wfge3'] = X['Wfgm3']/X['Wfga3']
    X['Lfge3'] = X['Lfgm3']/X['Lfga3']
    #Free Throw efficiency
    X['Wfte'] = X['Wftm']/X['Wfta']
    X['Lfte'] = X['Lftm']/X['Lfta']
    
    #Restrict to specified season if given
    if Season > 0:
        X = X[X['Season'] == Season]
            
    feature_dic = {}
    for i in range(len(raw_features)):
        feature_dic[['W'+feat for feat in raw_features][i]] = raw_features[i]
        feature_dic[['L'+feat for feat in raw_features][i]] = raw_features[i]   
    
    
    #Collect the stats for all games the team won
    X_W = X[(X.Wteam == team_ID)][['W'+feat for feat in raw_features]]
    X_W.rename(columns = feature_dic, inplace = True)
    X_W['Win'] = 1
    
    #Collect the stats for all games the team lost
    X_L = X[(X.Lteam == team_ID)][['L'+feat for feat in raw_features]]
    X_L.rename(columns = feature_dic, inplace = True)
    X_L['Win'] = 0

    #Combine win stats with loss stats and return
    return pd.concat([X_W, X_L])
```

```python
def matchup_stats(team_1, team_2, season):

    '''
    Input: two team_ID numbers and a season year
    Output: Dataframe containing the mean stats for each of the two teams for the given season. Incuded among these
        stats is the winrate for the team.
    
    Note: The output is insensitive to the order in which the two teams are specified.
    '''
    
    #By convention we decide to order the teams by team_ID
    A = min(team_1, team_2)
    B = max(team_1, team_2)
    
    #Get the stats for each team using team_stats
    A_stats = team_stats(A, season)
    B_stats = team_stats(B, season)
    
    #Rename the stats so that we can tell the team A stats apart from the team B stats in the final dataframe
    A_dic = {'Win' : 'Awrate'}
    for i in range(len(raw_features)):
        A_dic[raw_features[i]] = 'A'+raw_features[i]
    B_dic = {'Win' : 'Bwrate'}
    for i in range(len(raw_features)):
        B_dic[raw_features[i]] = 'B'+raw_features[i]
    
    A_stats.rename(columns = A_dic, inplace = True)
    B_stats.rename(columns = B_dic, inplace = True)
    
    #Return the mean of each teams stats
    return pd.concat([A_stats.mean(), B_stats.mean()])
```

```python
#tourn_features are the features we use to compare two teams in a given matchup.
tourn_features = ['A' + feat for feat in raw_features] + ['B' + feat for feat in raw_features]+\
    ['Awrate', 'Bwrate', 'Aseed', 'Bseed', 'Awin', 'Aorank', 'Borank']

def tourn_stats(season):
    
    '''
    Input: season year
    Output: Dataframe associated to the tournament for that year. Each row is a game played in that tournament,
        and the columns are the season statistics of the two teams which played in that game. One column says whether
        team A won the tournament matchup.
    '''
    
    
    tourn = pd.DataFrame(columns = tourn_features )
    
    results = tourn_results_detailed[tourn_results_detailed['Season'] == season]
    for i in results.index:
        #We continue the convention of making team A the one with lower team_ID
        A = min(results.ix[i].Wteam,results.ix[i].Lteam)
        B = max(results.ix[i].Wteam,results.ix[i].Lteam)

        #Gather the season statistics of the two teams.
        game_stats = pd.DataFrame(matchup_stats(A,B,season)).T
        
        #The seeds rankings and Massey ordinals are also used as features.
        game_stats['Aseed'] = int(tourn_seeds[(tourn_seeds.Season == season) &\
                                              (tourn_seeds.Team == A) ].Seed.values[0][1:3])
        game_stats['Bseed'] = int(tourn_seeds[(tourn_seeds.Season == season) &\
                                              (tourn_seeds.Team == B) ].Seed.values[0][1:3])
        
        #We take the mean of all of the various rankings in the Massey file. day_num==133 for the end of the season.
        game_stats['Aorank'] = massey[(massey.season == season) &\
                                      (massey.rating_day_num == 133) &\
                                      (massey.team == A)].mean().orank
        
        game_stats['Borank'] = massey[(massey.season == season) &\
                                      (massey.rating_day_num == 133) &\
                                      (massey.team == B)].mean().orank
        
    
        if A == results.ix[i].Wteam:
            game_stats['Awin'] = 1
        else:
            game_stats['Awin'] = 0 
        
        tourn = pd.concat([tourn, game_stats], ignore_index = True)
        
    return tourn
```

```python
#Compile statistics for all tournaments in a given range of years for data exploration.
explore = pd.DataFrame(columns = tourn_features )
for year in range(2003, 2016 +1):
    explore = pd.concat([explore,tourn_stats(year)], ignore_index=True)
```

```python
#raw_features = ['fge','fge3','fte','to','blk','stl','pf','ast','score', 'dr','or']
#standard ordering of the columns
features = [feat + 'Rat' for feat in raw_features] + [feat + 'Diff' for feat in raw_features] +\
['seedRat','seedDiff', 'wrateRat', 'wrateDiff', 'orankRat', 'orankDiff']

def compute_features(data):
    
    '''
    Input: Dataframe containing the matchup data for all of the games in a tournament, like the output of tourn_stats
    Output: Input datafame together with extra columns containing differences and logs of ratios of each of the stats
        between team A and team B.
        
    Taking the log of the ratios rather than just the ratios themselves seems to work better and avoids clumping of the
    data. Also, differences and ratios in general seem to be more efficient than just feeding the raw data itself
    into a model.
    '''
    
    for feat in raw_features:
        data[feat+'Rat'] = np.log(data['A'+feat]/data['B'+feat])
        data[feat+'Diff'] = data['A'+feat]-data['B'+feat]
    
    data['wrateRat'] = np.log(data.Awrate/data.Bwrate)
    data['wrateDiff'] = data.Awrate-data.Bwrate
    
    data['seedRat'] = np.log(data.Aseed/data.Bseed)
    data['seedDiff'] = data.Aseed-data.Bseed
    
    data['orankRat'] = np.log(data.Aorank/data.Borank)
    data['orankDiff'] = data.Aorank-data.Borank

    return data
```

## Training the Model

Now we compile all of the tourn_stats for a range of seasons to use as the training set for our model

```python
#Compile statistics for all tournaments in a given range of years
#Ultimately we will train the model on all years 2003-2016, but for testing purposes we train on 2003-2012
#and reserve 2013-2016 for testing
training_data = pd.DataFrame(columns = tourn_features )

for year in range(2003, 2016 +1):
    training_data = pd.concat([training_data,tourn_stats(year)], ignore_index=True)

training_data = compute_features(training_data)[features+['Awin']]

training_data.head()
```

This is where we define the model for calculating win probabilities. We are going to define a neural network with some number of nodes and let it run for some maximum number of iteractions before stopping. These paramters are determined by cross-validation. We randomly split the training data into training and test data, where the test_size is 25% and calculate a Log_Loss for both the training and test data. This is repeated 50 times for each value of the parameters.

Ideally we would simultaneously vary over both of parameters, but for simplicity we will do it one at a time. We begin with the number of nodes:

```python
X = training_data[features]
Y = training_data['Awin']

scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

#model = sklearn.naive_bayes.GaussianNB(priors = [0.5,0.5])
#model = sklearn.gaussian_process.GaussianProcessClassifier()
#model = sklearn.ensemble.RandomForestClassifier()
#model = sklearn.ensemble.AdaBoostClassifier()

train_losses_tot = []
test_losses_tot = []
node_range = range(5,26)
iter_num = 800
for node_num in node_range:
    model = sklearn.neural_network.MLPClassifier(activation='logistic',\
                                                 hidden_layer_sizes=(node_num,), tol = 1e-5, max_iter=iter_num)
    train_losses = []
    test_losses = []
    for i in range(50):
        X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y, test_size =0.25)
        model.fit(X_train,Y_train)
        #plt.plot(model.loss_curve_)
        #plt.show()
        train_losses.append(model.loss_)
        test_losses.append(-(Y_test.values.dot(np.log(model.predict_proba(X_test)[:,1]))\
           + (1-Y_test.values).dot(np.log(model.predict_proba(X_test)[:,0])))/len(Y_test))
        
    #print train_losses
    #print test_losses
    train_losses_tot.append(train_losses)
    test_losses_tot.append(test_losses)
        
#print train_losses_tot
#print test_losses_tot

plt.boxplot(train_losses_tot)
plt.title('Log_Loss for Training Set')
plt.xlabel('Number of Nodes')
plt.xticks(range(1,len(node_range)+1),node_range)
plt.show()
plt.boxplot(test_losses_tot)
plt.title('Log_Loss for Test Set')
plt.xlabel('Number of Nodes')
plt.xticks(range(1,len(node_range)+1),node_range)
plt.show()
```

The test data is relatively insensitive to the number of nodes for a fixed maximum number of iterations. We will use 15 nodes in the following. (Not because it's the most effiicient by some measure, but becase of historical reasons.) Note that this test also tells us we should expect the Log_Loss to lie somewhere in the range 0.55-0.60.

```python
X = training_data[features]
Y = training_data['Awin']

scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

#model = sklearn.naive_bayes.GaussianNB(priors = [0.5,0.5])
#model = sklearn.gaussian_process.GaussianProcessClassifier()
#model = sklearn.ensemble.RandomForestClassifier()
#model = sklearn.ensemble.AdaBoostClassifier()

train_losses_tot = []
test_losses_tot = []
iter_range = range(200,1401,100)
node_num = 15
for iter_num in iter_range:
    model = sklearn.neural_network.MLPClassifier(activation='logistic',\
                                                 hidden_layer_sizes=(node_num,), tol = 1e-5, max_iter=iter_num)
    train_losses = []
    test_losses = []
    for i in range(50):
        X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y, test_size =0.25)
        model.fit(X_train,Y_train)
        #plt.plot(model.loss_curve_)
        #plt.show()
        train_losses.append(model.loss_)
        test_losses.append(-(Y_test.values.dot(np.log(model.predict_proba(X_test)[:,1]))\
           + (1-Y_test.values).dot(np.log(model.predict_proba(X_test)[:,0])))/len(Y_test))
        
    #print train_losses
    #print test_losses
    train_losses_tot.append(train_losses)
    test_losses_tot.append(test_losses)
        
#print train_losses_tot
#print test_losses_tot

plt.boxplot(train_losses_tot)
plt.title('Log_Loss for Training Set')
plt.xlabel('Max Iterations')
plt.xticks(range(1,len(iter_range)+1),iter_range)
plt.show()
plt.boxplot(test_losses_tot)
plt.title('Log_Loss for Test Set')
plt.xlabel('Max Iterations')
plt.xticks(range(1,len(iter_range)+1),iter_range)
plt.show()
```

From this data we see that allowing too many iterations leads to overfitting. We choose 400 maximum iterations for the final model.Again we see that the expected Log_Loss is approximately in the range 0.55-0.60.

```python
X = training_data[features]
Y = training_data['Awin']

scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(X)
X = scaler.transform(X)


iter_num= 400
node_num = 15
model = sklearn.neural_network.MLPClassifier(activation='logistic',\
                                                 hidden_layer_sizes=(node_num,), tol = 1e-5, max_iter=400)
model.fit(X,Y)
plt.plot(model.loss_curve_)
plt.show()
```

## Generate Submission

This function generates all of the predictions for the possible matchups in a given tournament. The output is formatted according to the Kaggle specifications. The output of this function is the thing we submit to the competition.

```python
from itertools import combinations

def generate_predictions(season):
   
    ids = []
    pred = []
    
    matchups = tourn_seeds[tourn_seeds['Season']==season].Team
    matchups = combinations(matchups, 2)
    
    for n ,l in matchups:
        A = min(n,l)
        B = max(n,l)
        
        data = matchup_stats(A,B,season)
        
        data['Aseed'] = int(tourn_seeds[(tourn_seeds.Season == season) &\
                                              (tourn_seeds.Team == A) ].Seed.values[0][1:3])
        data['Bseed'] = int(tourn_seeds[(tourn_seeds.Season == season) &\
                                              (tourn_seeds.Team == B) ].Seed.values[0][1:3])
        
        if season < 2017:
            data['Aorank'] = massey[(massey.season == season) &\
                                      (massey.rating_day_num == 133) &\
                                      (massey.team == A)].mean().orank
        
            data['Borank'] = massey[(massey.season == season) &\
                                      (massey.rating_day_num == 133) &\
                                      (massey.team == B)].mean().orank
        else:
            data['Aorank'] = massey2017[(massey2017.season == season) &\
                                      (massey2017.rating_day_num == 133) &\
                                      (massey2017.team == A)].mean().orank
        
            data['Borank'] = massey2017[(massey2017.season == season) &\
                                      (massey2017.rating_day_num == 133) &\
                                      (massey2017.team == B)].mean().orank
        
        data = compute_features(data)[features]
        
        data = scaler.transform(data.values.reshape(1,-1))
        
        P = model.predict_proba(data)[0,1]
        
        ids.append('%s_%s_%s' %(season, A, B))
        pred.append(P)
        
    return ids, pred
```

```python
#Generate predictions for previous years for testing purposes and Stage 1 submission.
preds = pd.DataFrame(columns = ['id','pred'])
for year in range(2013, 2016 +1):
    ids, pred = generate_predictions(year)    
    temp_preds = pd.DataFrame(ids,columns = ['id'])
    temp_preds['pred'] = pred
    preds = pd.concat([preds,temp_preds])
preds.head()
```

```python
#Generate prediction for 2017 for the real competition.
ids, pred = generate_predictions(2017)
preds = pd.DataFrame(ids, columns = ['id'])
preds['pred'] = pred
preds.head()
```

## Stage 1 Testing (OLD)
This function will compute the log loss for a given set of predictions for a tournament from one of the seasons 2003-2016. The predictions are assumed to be in the form of the Kaggle submission file. A baseline value for the loss function is $\log 2 \approx 0.69$, which would be the result if one just flipped a coin for every game.

```python
def compute_loss(predictions, season):
    results = tourn_results_detailed[tourn_results_detailed['Season']== season]
    loss = 0
    for i in results.index:
        A = min(results.ix[i].Wteam,results.ix[i].Lteam)
        B = max(results.ix[i].Wteam,results.ix[i].Lteam)
        if A == results.ix[i].Wteam:
            loss += -np.log(predictions[predictions['id']== '%s_%s_%s' %(season,A,B)].pred.values[0])
        else:
            loss += -np.log(1-predictions[predictions['id']== '%s_%s_%s' %(season,A,B)].pred.values[0])
    return loss/len(results)
```

The first submission was not very sensible and was mostly an exercise in getting succssfully submittion something. The log loss is worse than random guessing.

```python
test = pd.read_csv('first_submission.csv')
loss = 0
for year in range(2013, 2016 +1):
    loss += compute_loss(test,year)
loss = loss/len(range(2013, 2016 +1))
loss
```
